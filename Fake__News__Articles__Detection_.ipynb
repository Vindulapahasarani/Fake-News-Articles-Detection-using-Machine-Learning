```python
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Download the necessary NLTK data
nltk.download('punkt')  # For tokenization
nltk.download('stopwords')  # For stopwords
nltk.download('wordnet')  # For lemmatization

# Initialize the WordNet lemmatizer
word_lem = WordNetLemmatizer()

def clean_data(text):
    words = []
    word_tokens = word_tokenize(text)
    for word in word_tokens:
        if word not in stopwords.words('english'):
            result = re.sub('[^a-zA-Z]', '', word)  # Remove non-alphabetical characters
            result = result.lower()  # Convert to lowercase
            result = word_lem.lemmatize(result)  # Lemmatize the word
            words.append(result)
    return ' '.join(words)  # Join the cleaned words

# Example usage
cleaned_text = clean_data('Hello, 123 Good Evening')
print(cleaned_text)
```

```python
articles_data['title'] = articles_data['title'].apply(clean_data)
```

```python
input_cols = articles_data['title']
output = articles_data['label']
```

```python
from sklearn.model_selection import train_test_split

# Assuming 'input_cols' and 'output' contain the features and labels respectively
x_train, x_test, y_train, y_test = train_test_split(input_cols, output, test_size=0.2, random_state=42)

# Now you can proceed with your training or other tasks
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
def predict(model_name, text):
  input_dataset =pd.DataFrame([text], columns=['title'])
  input_dataset['title'] = input_dataset['title'].apply(clean_data)
  text_vector = vector.transform(input_dataset['title'])
  result = model_name.predict(text_vector)
  if result[0] == 0 :
    return 'Its a Geniune Article'
  if result[0] == 1:
    return 'Its a Fake Artcile '
```
